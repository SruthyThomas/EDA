```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
data=pd.read_csv("myexcel.csv")
data
```

```python
##To check number of null values in the given dataset
data.isnull().sum()


```

```python

data.Salary.mode()
```

```python


newdata=data.fillna(data['Salary'].mode()[0])
newdata

        
```

# How many are there in each Team and the percentage splitting with respect to the total employees.

```python

count=newdata.Team.value_counts()
print(count)
names=count.index
sizes=count.values
plt.figure(figsize=(10,10))
plt.pie(sizes,labels=names,autopct='%.1f%%')
plt.title("Piechart")
plt.show()

```

# observations:
From the above piechart it is evident that more number of members present in  New Orleans Pelicans.Second position in terms of number of members is owned by Memphis Grizzlies  .

# Find if there is any correlation between age and salary , represent it visually

```python
newdata.Salary.value_counts()
```

```python
import matplotlib.ticker as ticker
x = newdata['Age']
y = newdata['Salary']  

# Create a scatterplot
plt.scatter(x, y)

# Define a function to format y-axis labels
def lakhs_millions_formatter(x, pos):
    if x < 1000000:
        return f'{x/100000:.0f}L'  # Format in lakhs
    else:
        return f'{x/1000000:.1f}M'  # Format in millions

# Apply the custom formatter to the y-axis
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lakhs_millions_formatter))
plt.title("Salary and Age Correlation")
plt.xlabel("Age")
plt.ylabel("Salary")

# Show the plot
plt.show()
```

```python
#finding correlation coefficient  between Age and Salary
newdata['Age'].corr( newdata['Salary'] ) 
```

# obsevations:
Correlation Coefficient is 0.20653501683058964.There is no correalation between Age and Salary.
